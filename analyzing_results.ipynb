# Analyzing results

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib as plt

plt.rcParams['figure.figsize'] = [8.0, 6.0]
plt.rcParams['figure.dpi'] = 100
plt.rcParams['savefig.dpi'] = 120
```

## Reading training results from CSV files

```python
results_awd_lstm_word = pd.read_csv('./results/awd_lstm_word.csv')
results_awd_lstm_unigram = pd.read_csv('./results/awd_lstm_unigram.csv')
results_awd_lstm_bpe = pd.read_csv('./results/awd_lstm_bpe.csv')
results_awd_lstm_char = pd.read_csv('./results/awd_lstm_char.csv')
results_awd_qrnn_word = pd.read_csv('./results/awd_qrnn_word.csv')
results_awd_qrnn_unigram = pd.read_csv('./results/awd_qrnn_unigram.csv')
results_awd_qrnn_bpe = pd.read_csv('./results/awd_qrnn_bpe.csv')
results_awd_qrnn_char = pd.read_csv('./results/awd_qrnn_char.csv')


results_gpt2 = pd.read_csv('./results/gpt2.csv')
results_bert = pd.read_csv('./results/bert.csv')
results_roberta = pd.read_csv('./results/roberta.csv')

```

```python
display(results_awd_lstm_word)
```

## Comparing accuracy over the training epochs

```python
ax = results_awd_lstm_word.plot(x='epoch', y='accuracy')
ax = results_awd_lstm_unigram.plot(ax=ax, x='epoch', y='accuracy')
ax = results_awd_lstm_bpe.plot(ax=ax, x='epoch', y='accuracy')
ax = results_awd_lstm_char.plot(ax=ax, x='epoch', y='accuracy')

ax.legend(["AWD-LSTM-word", 
           "AWD-LSTM-unigram", 
           "AWD-LSTM-bpe", 
           "AWD-LSTM-char"], 
          loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=4)

ax.set_xlabel("epochs")
ax.set_ylabel("accuracy");
```

```python
ax = results_awd_qrnn_word.plot(x='epoch', y='accuracy')
ax = results_awd_qrnn_unigram.plot(ax=ax, x='epoch', y='accuracy')
ax = results_awd_qrnn_bpe.plot(ax=ax, x='epoch', y='accuracy')
ax = results_awd_qrnn_char.plot(ax=ax, x='epoch', y='accuracy')

ax.legend(["AWD-QRNN-word", 
           "AWD-QRNN-unigram", 
           "AWD-QRNN-bpe", 
           "AWD-QRNN-char"], 
          loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=4)

ax.set_xlabel("epochs")
ax.set_ylabel("accuracy");
```
